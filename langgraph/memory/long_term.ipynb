Now using embeddings method which is easier to implement

```python
import os 
os.environ["HF_HOME"] = "E:\\huggingface_embedding_cache"
from langchain_huggingface import HuggingFaceEmbeddings

model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2",
    encode_kwargs={"normalize_embeddings": True},
)

store = InMemoryStore(index={
    "dims": 1536,
    "embed": model
})

namespace = ('user', 'nadeem')

store.put(namespace, 
          'favourite Sport',
          {"data": "Nadeem likes to play cricket"}
        )
store.put(namespace, 
          'hobby',
          {"data": "Nadeems hobby is gym"}
        )
store.put(namespace, 
          'food',
          {"data": "Nadeem likes to eat eggs"}
        )
store.put(namespace, 
          'today',
          {"data": "Nadeem will watch football final today"}
        )

memories = store.search(namespace, query="what user will do today and he will eat what he loves", limit=3)
for memory in memories:
    print(memory)
```
